```python
import numpy as np

arr = np.array([1, 3, 5, 7])
print('numpy_vector_array:', arr) # 1D vector array

arr.shape # to know shape of array
```

```python
matrix = np.array([[1, 3, 5, 7], [2, 4, 6, 8]])
print('numpy_matrix_array:', matrix) # 2D matrix array

matrix.shape  # to know shape of array
```

```python
# dtype() -> to know type of data

a = np.array([1, 2, 3])
b = np.array([1.5, 2.3, 3.7])

print('type a:', a.dtype)
print('type b:', b.dtype)
```

```python
# reshape() -> to reshape of data structure

a = np.array([1, 2, 3, 4, 5, 6])
b = a.reshape(2, 3)

print("shaped_array:", b)
print("original_array:", a)


# Important:
# `reshape` qilayotganda elementlar soni mos bo‘lishi kerak.
```

```python
# astype() -> change data type

a = np.array([1, 2, 3])
b = a.astype(float)

print('float:', b)
print('original:', a)
```

```python
# np.zeros() -> faqat 0 lardan iborat array yaratadi
# np.ones() -> faqat 1 lardan iborat array yaratadi

zeros = np.zeros((2, 3))
ones = np.ones((2, 3))

print("zeros:", zeros)
print("ones:", ones)
```

```python
# ndim() -> array necha o‘lchamli ekanligini ko‘rsatadi

arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr.ndim)
# 2

# 1D -> oddiy array
# 2D -> matrix
# 3D -> ko‘p o‘lchamli array
```

```python
# arange() -> berilgan oraliqda ketma-ket sonlardan array yaratadi

arr = np.arange(1, 6)
arr = np.arange(0, 10, 2)

print('arange:', arr)
print('arange with step:', arr)
```

```python
# sum() -> array ichidagi barcha sonlarni qo‘shadi
# mean() -> array ichidagi sonlarning o‘rtacha qiymatini chiqaradi
# min() -> eng kichik qiymatni topadi
# max() -> eng katta qiymatni topadi

arr = np.array([1, 2, 3, 4])

print(arr.sum())
print(arr.mean())
print(arr.min())
print(arr.max())
```

```python
# flatten() -> ko‘p o‘lchamli array’ni 1D array’ga aylantiradi

matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

new_arr = matrix.flatten()

print(new_arr)
# [1 2 3 4 5 6]
```

```python
# T (Transpose) -> matrix’ning qator va ustunlarini almashtiradi

matrix = np.array([
    [1, 2],
    [3, 4]
])

new_matrix = matrix.T

print(new_matrix)
```

```python
# dot() -> ikki array/matrix orasida dot product hisoblaydi

a = np.array([
      [1, 2],
      [3, 4]
    ])

b = np.array([
      [4, 6],
      [7, 8]
    ])


print(a.dot(b))

# row from first matrix × column from second matrix
# calculation:
# [1, 2] × [4, 7] = 18
# [1, 2] × [6, 8] = 22
# [3, 4] × [4, 7] = 40
# [3, 4] × [6, 8] = 50
```

```python
# random.rand() -> 0 va 1 oralig‘ida random float sonlar yaratadi

arr = np.random.rand(2, 3)

print(arr)
```

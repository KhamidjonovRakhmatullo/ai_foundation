```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("StudentsPerformance.csv")

print("\n==========================================\n")

print("Dataset shape (rows, columns):", df.shape)

print("\n==========================================\n")

print("Column name:", df.columns.tolist())

print("\n==========================================\n")

print("First 5 rows of the dataset:")
print(df.head())

print("\n==========================================\n")

print("Summary statistics:")
print(df.describe())

print("\n==========================================\n")

print("Data types and null info:")
print(df.info())

print("\n==========================================\n")
```

```python
print("==========================================\n")

print("Avarage scores by subjects:")
df[["math score", "reading score", "writing score"]].mean()
```

```python
print("==========================================\n")

print("Maximum scores by subjects:")
df[["math score", "reading score", "writing score"]].max()
```

```python
print("==========================================\n")

print("Min scores by subjects:")
df[["math score", "reading score", "writing score"]].min()
```

```python
# Correlation
print("==========================================\n")

print("Correlation:")
df[["math score", "reading score", "writing score"]].corr()
```

```python
print("==========================================\n")

# total score -> math, reading, writing score'larni qo‘shadi
# axis=1 -> row bo‘yicha hisoblaydi, ya’ni har bir student uchun alohida qo‘shadi
df["total score"] = df[["math score", "reading score", "writing score"]].sum(axis=1)


# average score -> total score'ni 3 ga bo‘lib, o‘rtacha ballni hisoblaydi
df["average score"] = df["total score"] / 3


# passed -> agar average score 50 yoki undan katta bo‘lsa "Yes", aks holda "No"
# np.where(condition, true_value, false_value)
df["passed"] = np.where(df["average score"] >= 50, "Yes", "No")


# faqat yangi yaratilgan column'larning birinchi 5 ta row'ini ko‘rsatadi
print(df[["total score", "average score", "passed"]].head())
```

```python
print("==========================================\n")

# Preparation-based average

# groupby() -> data'ni berilgan column bo‘yicha guruhlaydi
# bu yerda students "test preparation course" bo‘yicha guruhlanadi

test_prep_group = df.groupby("test preparation course")[["math score", "reading score", "writing score"]].mean()

# "test preparation course" bo‘yicha:
# math score, reading score, writing score column'larining o‘rtacha qiymatini chiqaradi


print("student score by test preparation course: \n")
print(test_prep_group)
# result:
# completed -> test preparation course qilgan student'larning o‘rtacha ballari
# none      -> test preparation course qilmagan student'larning o‘rtacha ballari


print("==========================================\n")

# Gender-based average

gender_avg = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()

print("student score by gender: \n")
print(gender_avg)


print("==========================================\n")

# Passed vs Failed

# value_counts() -> "passed" column ichida Yes va No nechta ekanini sanaydi
# normalize=True -> son o‘rniga foiz/proportion qilib chiqaradi
# * 100 -> proportion'ni percentage'ga aylantiradi

pass_rate = df["passed"].value_counts(normalize=True) * 100


print("student pass rate: \n")

print(pass_rate)
# Yes -> passed student'lar foizi
# No  -> failed student'lar foizi
```

```python
print("==========================================\n")

print("Current score columns in data frame: \n")
print(df[["math score", "reading score", "writing score"]].head())


print("\n==========================================\n")


# Convert scores to NumPy array

scores = df[["math score", "reading score", "writing score"]].values
# Pandas DataFrame'dagi 3 ta score column'ni NumPy array/matrix ko‘rinishiga o‘tkazadi

print("\n NumPy Array Matrix:")
print(scores)


print("\n==========================================\n")


# Mean and Standard Deviation

mean_scores = np.mean(scores, axis=0)
# axis=0 -> har bir column bo‘yicha hamma 1000 row'larning mean hisoblaydi (average)
# math score mean, reading score mean, writing score mean

print("\nNumPy Mean: \n")
print(mean_scores)

print("\n---")

std_scores = np.std(scores, axis=0)
# axis=0 -> har bir column bo‘yicha hamma 1000 row'larning std hisoblaydi (bir-biriga qanchalik o'xshash(yaqin yoki uzoq) result'lar)
# har bir score column qanchalik tarqalganini ko‘rsatadi

print("\nNumPy Std: \n")
print(std_scores)


print("\n==========================================\n")


# Z-score / Normalization

z_score = (scores - mean_scores) / std_scores
# har bir score'dan o‘sha column mean'ini ayiradi
# keyin std'ga bo‘ladi
# natijada score'lar normalized bo‘ladi

print("\nZ-scores / Normalized numbers:\n")
print(z_score[:5])
# faqat birinchi 5 ta row'ni chiqaradi

# z-score around 0  -> average score'ga yaqin
# z-score negative  -> average'dan past
# z-score positive  -> average'dan yuqori
```

```python
# ------------------------------ Student performance check ------------------------------

# 1) How many students scored 90 or above average?

# average score 90 yoki undan yuqori bo‘lgan student'larni filter qiladi
top_scorers = df[df["average score"] >= 90]

print("Top scored students above 90:", len(top_scorers))


print("\n==========================================\n")


# 2) How many students failed?

# passed column'i "No" bo‘lgan student'larni filter qiladi
failed = df[df["passed"] == "No"]

print("Failed from exam:", len(failed))
```

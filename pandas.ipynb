```python
# ------------------------------ iloc ------------------------------

# iloc -> row va column'larni index raqami orqali tanlaydi
# iloc = integer location


# 1) Birinchi row'ni chiqarish

df.iloc[0]
# birinchi row chiqadi


# ---------------------------------------------------------------------

# 2) Birinchi 5 ta row'ni chiqarish

# df.iloc[0:5]
# 0-indexdan 5-indexgacha chiqaradi
# 5-index kirmaydi


# ---------------------------------------------------------------------

# 3) 0-indexdan tashqari hamma row'ni chiqarish

# df.iloc[1:]
# 1-indexdan boshlab oxirigacha chiqaradi


# ---------------------------------------------------------------------

# 4) Bitta qiymatni olish

# df.iloc[0, 1]
# 0-indexdagi row va 1-indexdagi column qiymatini chiqaradi


# ---------------------------------------------------------------------

# 5) Faqat birinchi column'ni chiqarish

# df.iloc[:, 0]
# barcha row'lar, faqat 0-indexdagi column chiqadi


# ---------------------------------------------------------------------

# 6) Birinchi 5 ta row va birinchi 2 ta column'ni chiqarish

# df.iloc[0:5, 0:2]
# 0 dan 5 gacha row, 0 dan 2 gacha column
# 5 va 2 kirmaydi


# ---------------------------------------------------------------------

# 7) Oxirgi row'ni chiqarish

# df.iloc[-1]
# oxirgi row chiqadi


# ---------------------------------------------------------------------

# Simple meaning:

# df.iloc[0]          -> first row
# df.iloc[0:5]        -> first 5 rows
# df.iloc[1:]         -> all rows except first row
# df.iloc[0, 1]       -> row 0, column 1 value
# df.iloc[:, 0]       -> first column
# df.iloc[0:5, 0:2]   -> first 5 rows and first 2 columns
# df.iloc[-1]         -> last row
```

```python
# Missing value bor dataset yaratamiz

data_with_missing = {
  'cars': ["BMW", "Volvo", None, "Tesla", "Ferrari", None, "Bugatti"],
  'counts': [3, 7, None, 8, 1, 3, None]
}

df_missing = pd.DataFrame(data_with_missing)

print(df_missing)
```

```python
# ------------------------------ isnull() ------------------------------

# isnull() -> missing value bor joylarni True/False qilib ko‘rsatadi


# 1) Missing value bor joylarni ko‘rish

df_missing.isnull()
# True  -> missing value bor
# False -> missing value yo‘q


# ---------------------------------------------------------------------

# 2) Har bir column'da nechta missing value borligini sanash

# df_missing.isnull().sum()
# cars va counts column'larida nechta NaN/None borligini chiqaradi


# ---------------------------------------------------------------------

# 3) Umumiy nechta missing value borligini sanash

# df_missing.isnull().sum().sum()
# butun DataFrame ichidagi jami missing value sonini chiqaradi


# ---------------------------------------------------------------------

# 4) Bitta column bo‘yicha missing value tekshirish

# df_missing["cars"].isnull()
# faqat cars column'idagi missing joylarni ko‘rsatadi


# df_missing["counts"].isnull()
# faqat counts column'idagi missing joylarni ko‘rsatadi


# ---------------------------------------------------------------------

# 5) Missing value bor row'larni chiqarish

# df_missing[df_missing["cars"].isnull()]
# cars column'ida missing value bor row'larni chiqaradi


# df_missing[df_missing["counts"].isnull()]
# counts column'ida missing value bor row'larni chiqaradi


# ---------------------------------------------------------------------

# Simple meaning:

# isnull()              -> missing joylarni True/False qiladi
# isnull().sum()        -> har bir column'dagi missing sonini chiqaradi
# isnull().sum().sum()  -> jami missing value sonini chiqaradi
```

```python
# ------------------------------ dropna() ------------------------------

# dropna() -> missing value bor row'larni olib tashlaydi

# ---------------------------------------------------------------------

# 1) Missing value bor row'larni olib tashlash

df_missing.dropna()
# bitta NaN/None bo‘lsa ham o‘sha row o‘chadi


# ---------------------------------------------------------------------

# 2) Faqat hamma qiymatlari missing bo‘lgan row'larni olib tashlash

# df_missing.dropna(how="all")
# row ichidagi barcha column'lar NaN/None bo‘lsa, o‘sha row o‘chadi


# ---------------------------------------------------------------------

# 3) Missing value bor column'larni olib tashlash

# df_missing.dropna(axis=1)
# qaysi column'da NaN/None bo‘lsa, o‘sha column o‘chadi


# ---------------------------------------------------------------------

# 4) Faqat bitta column bo‘yicha tekshirish

# df_missing.dropna(subset=["cars"])
# cars column'ida missing value bor row'larni o‘chiradi


# df_missing.dropna(subset=["counts"])
# counts column'ida missing value bor row'larni o‘chiradi


# ---------------------------------------------------------------------

# 5) inplace=True -> original DataFrame'ni o‘zgartiradi

# df_missing.dropna(inplace=True)
# natija yangi variable'ga saqlanmaydi, df_missing o‘zi o‘zgaradi


# ---------------------------------------------------------------------

# Simple meaning:

# dropna()                 -> missing bor row'larni o‘chiradi
# dropna(how="all")        -> hammasi missing bo‘lgan row'ni o‘chiradi
# dropna(axis=1)           -> missing bor column'larni o‘chiradi
# dropna(subset=["cars"])  -> faqat cars column bo‘yicha tekshiradi
# inplace=True             -> result'ni original DataFrame'ga saqlaydi
```

```python
# ------------------------------ fillna() ------------------------------

# fillna() -> missing value'larni boshqa qiymat bilan to‘ldiradi

# ---------------------------------------------------------------------

# 1) Barcha missing value'larni bitta qiymat bilan to‘ldirish

df_missing.fillna("Unknown")
# barcha NaN/None qiymatlarni "Unknown" bilan to‘ldiradi


# ---------------------------------------------------------------------

# 2) Har xil column'larni har xil qiymat bilan to‘ldirish

# df_missing.fillna({
#     "cars": "Unknown",
#     "counts": 0
# })

# cars   -> None qiymatlar "Unknown" bo‘ladi
# counts -> NaN qiymatlar 0 bo‘ladi


# ---------------------------------------------------------------------

# 3) Faqat bitta column bo‘yicha to‘ldirish

# df_missing["cars"].fillna("Unknown")
# cars column'dagi missing qiymatlarni "Unknown" qiladi


# df_missing["counts"].fillna(0)
# counts column'dagi missing qiymatlarni 0 qiladi


# ---------------------------------------------------------------------

# 4) Mean bilan to‘ldirish

# df_missing["counts"].fillna(df_missing["counts"].mean())
# counts column'dagi missing qiymatlarni o‘rtacha qiymat bilan to‘ldiradi


# ---------------------------------------------------------------------

# 5) inplace=True -> original DataFrame'ni o‘zgartiradi

# df_missing.fillna({
#     "cars": "Unknown",
#     "counts": 0
# }, inplace=True)


# ---------------------------------------------------------------------

# Simple meaning:

# fillna("Unknown") -> missing joylarni "Unknown" bilan to‘ldiradi
# fillna(0)         -> missing joylarni 0 bilan to‘ldiradi
# fillna(mean)      -> missing joylarni o‘rtacha qiymat bilan to‘ldiradi
# inplace=True      -> result'ni original DataFrame'ga saqlaydi
```
